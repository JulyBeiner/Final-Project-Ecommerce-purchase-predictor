identificamos que algunas columnas del dataset están codificadas numéricamente. Para facilitar la interpretación y mejorar la visualización en el análisis exploratorio de datos (EDA), asignaremos nombres descriptivos a estas columnas mediante un proceso de mapeo. Esto nos permitirá analizar los datos de manera más clara y comprensible.

```python
browser_mapping = {
    1: "Chrome",
    2: "Safari",
    3: "Internet Explorer",
    4: "Firefox",
    5: "Edge",
    6: "Opera",
    7: "Otros",
    8: "Otros",
    9: "Otros",
    10: "Otros",
    11: "Otros",
    12: "Otros",
    13: "Otros"
}
```

```python
df['Browser_Label'] = df['Browser'].map(browser_mapping)
```

```python
os_mapping = {
    1: "Windows",
    2: "MacOS",
    3: "Linux",
    4: "Android",
    5: "iOS",
    6: "Other",
    7: "Other",
    8: "Other"
}
```

```python
df['OS_Label'] = df['OperatingSystems'].map(os_mapping)
```

```python
df.Region.unique()
```

```python
region_mapping = {
    1: "NorthAmerica",
    2: "Europe",
    3: "Asia",
    4: "SouthAmerica",
    5: "Africa",
    6: "Oceania",
    7: "MiddleEast",
    8: "Antarctica",
    9: "Other"
}
```

```python
df['Region_Label'] = df['Region'].map(region_mapping)
```

```python
df.TrafficType.nunique()
```

```python
traffic_mapping = {
    1: "Organic Traffic",
    2: "Direct Traffic",
    3: "Referral Traffic",
    4: "Paid Traffic",
    5: "Social Media",
    6: "Email Marketing",
    7: "Other",
    8: "Other",
    9: "Other",
    10: "Other",
    11: "Other",
    12: "Other",
    13: "Other",
    14: "Other",
    15: "Other",
    16: "Other",
    17: "Other",
    18: "Other",
    19: "Other",
    20: "Other"
}
```

```python
df['Traffic_Type_Label'] = df['TrafficType'].map(traffic_mapping)
```

```python
df['VisitorType'] = df['VisitorType'].str.replace('_', ' ')
```

```python
column_mapping = {
    "ProductRelated": "Product_Related",
    "ProductRelated_Duration": "Product_Related_Duration",
    "BounceRates": "Bounce_Rates",
    "ExitRates": "Exit_Rates",
    "PageValues": "Page_Values",
    "SpecialDay": "Special_Day",
    "OperatingSystems": "Operating_Systems",
    "TrafficType": "Traffic_Type",
    "VisitorType": "Visitor_Type"
}
```

```python
df.rename(columns=column_mapping, inplace=True)
```

```python
df.columns = [col.capitalize() for col in df.columns]
```
